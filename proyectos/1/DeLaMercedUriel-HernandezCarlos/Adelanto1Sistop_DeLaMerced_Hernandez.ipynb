Para el formato de la memoria, nos apoyamos de referencias, como el simulador del microprocesador Z80 utilizado el anterior semestre, diseñado por Virueña Luis (foto adjuntada) el cual muestra el nombre de las filas y columnas en hexadecimal, además de presentar el caractér a la derecha de su representación en código ASCII.

```python
def print_memory(memory):
    # Función para imprimir la representación de la memoria en formato hexadecimal
    print("  ", end="")
    # Imprime los encabezados de columna en formato hexadecimal
    for i in range(16):
        print(f"{i:02X} ", end="")
    print()
    # Itera sobre cada fila de la memoria e imprime sus valores hexadecimales
    for i, row in enumerate(memory):
        # Imprime el número de fila en formato hexadecimal
        print(f"{i:04X} ", end="")
        # Itera sobre cada valor hexadecimal en la fila y lo imprime
        for hex_val in row:
            print(f"{hex_val:02X} ", end="")
        print()

def main():
    # Inicializa la memoria con ceros
    memory = [[0 for _ in range(16)] for _ in range(16)]

    while True:
        print_memory(memory)  # Imprime la representación actual de la memoria
        # Solicita al usuario que ingrese una dirección en formato hexadecimal
        print("\nIngrese fila y columna (por ejemplo, '000C' para fila 0, columna C), o 'q' para salir:")
        address = input().lower()

        if address == 'q':
            break  # Sale del bucle si el usuario ingresa 'q' para salir del programa

        try:
            row = int(address[:4], 16)  # Obtiene la fila de la dirección ingresada
            col = int(address[4:], 16)  # Obtiene la columna de la dirección ingresada
            if 0 <= row < 16 and 0 <= col < 16:  # Verifica si la dirección está dentro del rango de la memoria
                # Solicita al usuario que ingrese un nuevo valor hexadecimal para esa dirección
                print(f"Ingrese nuevo valor hexadecimal para la dirección {address} (0-F):")
                new_val = input().lower()
                if new_val.isdecimal() and int(new_val) in range(16):  # Verifica si el valor ingresado es válido
                    memory[row][col] = int(new_val, 16)  # Actualiza el valor en la memoria
                else:
                    # Imprime un mensaje de error si el valor no es válido
                    print("Entrada inválida. Por favor, ingrese un valor hexadecimal válido (0-F).")
            else:
                # Imprime un mensaje de error si la dirección no es válida
                print("Entrada inválida. Por favor, ingrese una dirección hexadecimal válida en el formato '000C'.")
        except ValueError:
            # Imprime un mensaje de error si la entrada no es válida
            print("Entrada inválida. Por favor, ingrese una dirección hexadecimal válida en el formato '000C'.")

if __name__ == "__main__":
    main()  # Llama a la función principal si el script se ejecuta directamente
```
